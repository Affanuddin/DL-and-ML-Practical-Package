```python
movie_correlations = user_id_matrix.corr(method =  'pearson', min_periods = 80)
```

```python
movie_correlations
```

```python
movie_data['Movie Name'][0]
```

```python
similar_movie_list = pd.Series()

for i in range(0,2):
    similar_movie = movie_correlations[   movie_data['Movie Name'][i]  ].dropna()
    similar_movie = similar_movie.map(lambda x: x*movie_data['Ratings'][i])
    similar_movie_list = similar_movie_list.append(similar_movie)
```

```python
similar_movie_list.sort_values(inplace=True, ascending=False)
print(similar_movie_list.head(10))
```
